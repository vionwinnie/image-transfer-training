```python
import pandas as pd
```

```python
df = pd.read_csv('../raw/file_name.csv',index_col=0)
df.head()
```

```python
df.classification.value_counts()
```

```python
pos_df = df.loc[df.classification ==1,:]
neg_df = df.loc[df.classification == 0,:]
print(pos_df.shape)
print(neg_df.shape)

```

```python
neg_df_short =  neg_df.sample(frac = 0.3)
```

```python
# Creating a dataframe with 75% values of original dataframe
def split_train_val(df):
    part_75 = df.sample(frac = 0.75)
    part_25 = df.drop(part_75.index)
    return part_75, part_25
```

```python
#pos_df_train, pos_df_val = split_train_val(pos_df)
neg_df_train, neg_df_val = split_train_val(neg_df_short)
print(neg_df_train.shape, neg_df_val.shape)
```

```python
import shutil 

cur_df = pos_df_train
folder = "plate"
splits = "train"

for idx, row in cur_df.iterrows():
    cur_file_name = row.file_name.replace("\n","")
    cur_file_path = "../raw/"+ cur_file_name
    cur_file_raw_name = cur_file_name.split("/")[-1]
    cur_export_path = "/home/winnie/food-pics/labels/{}/{}/{}".format(splits,folder,cur_file_raw_name)
    
    print(cur_file_path)
    print(cur_export_path)
    shutil.copy(cur_file_path,cur_export_path)
    
```
